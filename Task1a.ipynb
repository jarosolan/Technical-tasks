```python
def function():
    """This function returns the first number after the 40755, 
    which is triangular, pentagonal and hexagonal simultaneously"""
    # Create 3 lists to store triangular, pentagonal and hexagonal numbers
    T = [] # List for triangular numbers
    P = [] # List for pentagonal numbers
    H = [] # List for hexagonal numbers
    number = 0
    # Calculating all triangular numbers, just after the index of last found number (285)
    for i in range(286, 100000): # Setting range for loop to calculate values
        n = int(i*(i+1)/2)       # Calculating value
        T.append(n)              # Adding result to list
    # Calculating all pentagonal numbers, just after the index of last found number (165)
    for i in range(166, 50000):
        n = int(i*(3*i-1)/2)
        P.append(n)
    # Calculating all hexagonal numbers, just after the index of last found number (143)
    for i in range(144, 50000):
        n = i*(2*i-1)
        H.append(n)
    # Storing results as sets to intersect each other to find the triangle number
    set_T = set(T)
    set_P = set(P)
    set_H = set(H)
    # Intersecting 3 sets together
    N = set_T.intersection(set_P).intersection(set_H)
    # Give the final result
    print("The next triangle number is:", N)
function()
```

```python
def function2():
    """This function return number which is triangular, pentagonal and hexagonal"""
    # Create 2 lists to store pentagonal and hexagonal numbers
    P = []
    H = []
    # Calculating all pentagonal numbers, just after the index of last found number (165)
    for i in range(166, 50000):
        n = int(i*(3*i-1)/2)
        P.append(n)
    # Calculating all hexagonal numbers, just after the index of last found number (143)
    for i in range(144, 50000):
        n = i*(2*i-1)
        H.append(n)
    # Storing results as sets to intersect them to find the number
    set_P = set(P)
    set_H = set(H)
    # Intersecting 2 sets together
    number = set_P.intersection(set_H)
    # Give the final result
    print("The next triangle number is:", number)
function2()
```
